```python
# 문평동 pm10과 기온

# scatter chart

# figure size 지정
plt.figure(figsize=(5, 10))

plt.plot(pm10['Avg_tem'],pm10['Munpyung'], 'o', color='b', alpha=0.5)


# label과 title 정하기
plt.ylabel('pm10_Munp', fontsize=30)
plt.xlabel('Avg_tem',fontsize=30)

# 회귀선
pm10_Munpyung_tem = sns.lmplot(x="Avg_tem", y="Munpyung", data = pm10)

plt.show()
```
