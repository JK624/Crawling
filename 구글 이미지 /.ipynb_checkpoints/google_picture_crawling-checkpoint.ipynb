```python
import sys # 시스템
import os  # 시스템

import pandas as pd  # 판다스 : 데이터분석 라이브러리
import numpy as np   # 넘파이 : 숫자, 행렬 데이터 라이브러리

from bs4 import BeautifulSoup    # html 데이터를 전처리

from selenium import webdriver  # 라이브러리(모듈) 가져오라
from selenium.webdriver import ActionChains as AC
from tqdm import tqdm # 진행표시바

#from tqdm import tqdm_notebook # 진행표시바(사용시 실행은 되는 오류 발생)
from time import sleep
import time

from selenium.webdriver.common.keys import Keys # 사용해야 엔터키 사용 가능

# 이미지들 image_google 폴더에 다운받기
import urllib.request

dict = {}                                   # 전체 데이터를 담을 딕셔너리 생성

# *************************************라이브러리 선언***************************** #
# 스크롤을 밑으로 내려주는 함수
def scroll_down(driver):
    driver.execute_script("window.scrollTo(0,100000)")
    time.sleep(1)

# *************************************함수 선언*********************************** #
```

```python
while(1):

    keyword = input('구글에 검색 할 키워드를 입력해주세요.')
    print(keyword)

# ************************************키워드 선언********************************** #

    path = "/Users/jisookwon/Documents/chromedriver"       # 윈도우는 "chromedriver.exe"

    driver = webdriver.Chrome(path)
    driver.get('https://www.google.com/')
    time.sleep(1)

# ***********************************크롬 실행************************************* #

    # 검색창에 커서 클릭
    search = driver.find_element_by_css_selector(".gLFyf.gsfi")
    search.click()
    time.sleep(1)

    # 검색 창에 검색어 입력
    search = driver.find_element_by_css_selector(".gLFyf.gsfi")
    search.send_keys(keyword)
    time.sleep(3)

    # 검색어 입력후 엔터
    search.send_keys(Keys.ENTER)
    time.sleep(1)

    # 구글 이미지 검색 클릭
    # driver.find_element_by_css_selector(".hide-focus-ring").click()
    driver.find_element_by_link_text("이미지").click( )

    # 스크롤 다운 (구글에서 한번에 보여주는 사진의 수는 500개)
    n = 5         # 스크롤 횟수
    i = 0
    while i < n: # 이 조건이 만족되는 동안 반복 실행
        scroll_down(driver) # 스크롤 다운
        i = i+1

# ****************************크롬 스크롤 화면 확장******************************** #
    # 사진 클릭
    ft_pic = "#islrg > div.islrc > div:nth-child(1) > a.wXeWr.islib.nfEiy.mM5pbd > div.bRMDJf.islir > img"
    ft_pic_button = driver.find_element_by_css_selector(ft_pic)     # 사진 버튼 찾기
    ft_pic_button.click()       # 사진 버튼 클릭
    time.sleep(2)

# ***********************************사진 클릭************************************* #

    # 크롤링 시작
    len_google = n*100   # 몇 개 글을 크롤링 할건지

    # for i in tqdm_notebook(range(0, 5)):   # 실행은 되는 오류발생(아래와 같음)
    for i in tqdm(range(0, 5)):    # range : 숫자가 1씩하는 함수

        target_info = {}                      # 사진별 데이터를 담을 딕셔너리 생성

        try:    # 사진(pic) 크롤링 시도
            # overlays1 = ".n3VNCb"  # 사진창 속 사진 , 이유는 모르겠지만 클래스 명으로 할 경우 
                                     # i가 1이상일때부터 i-1값의 사진이 저장됨                                     
            overlays1 = "#Sva75c > div > div > div.pxAole > div.tvh9oe.BIB1wf > c-wiz > div > div.OUZ5W > div.zjoqD > div > div.v4dQwb > a > img"
            # 링크 주소를 selector로 사용하면 원하는 결과를 도출할 수 있음
            
            img = driver.find_element_by_css_selector(overlays1)   # 사진 선택
            pic = img.get_attribute('src')                         # 사진 url 크롤링 완료
            target_info['picture'] = pic
            
            dict[i] = target_info            # 토탈 딕셔너리로 만들기
    
            # 다음장 클릭
            CSS_tran2="#Sva75c > div > div > div.pxAole > div.tvh9oe.BIB1wf > c-wiz > div > div.OUZ5W > div.zjoqD > div > div.ZGGHx > a.gvi3cf > div"             # 다음 버튼 정의
            tran_button2 = driver.find_element_by_css_selector(CSS_tran2)  # 다음 버튼 find
            tran_button2.click()

        except:  # 에러가 나면 다음장을 클릭해라  
            CSS_tran2="#Sva75c > div > div > div.pxAole > div.tvh9oe.BIB1wf > c-wiz > div > div.OUZ5W > div.zjoqD > div > div.ZGGHx > a.gvi3cf > div"             # 다음 버튼 정의
            tran_button2 = driver.find_element_by_css_selector(CSS_tran2)  # 다음 버튼 find
            tran_button2.click()     # 다음 버튼 클릭
            time.sleep(2)

    print(dict)

# ***********************************사진 크롤링*********************************** #

    # 판다스로 만들기 : 엑셀(테이블) 형식으로 만들기
    result_df = pd.DataFrame.from_dict(dict, 'index')

    n = result_df['picture'].count()

    # csv 형식(엑셀이랑 비슷한 파일)으로 저장
    result_df.to_csv("google_({}).csv".format(keyword), encoding='euc-kr')

# *********************************엑셀 만들기************************************* #

    # 만약 image_google 폴더가 없으면 만들어라
    if not os.path.exists("image_google"):
        os.makedirs("image_google")

    for i in range(0, n):

        try:
            index = result_df['picture'][i]
            urllib.request.urlretrieve(index, "image_google/{0}{1}.jpg".format(keyword,i))        
        except:
            pass
        
        
    driver.close() # 실행중인 크롬 창 닫기
    
# ***********************************사진 다운************************************* #
    
    replay = input('다른것을 검색하시겠습니까? yes(1) or no(0)     ')
    print(replay)
    if(replay == 'no' or replay == '0'):
        break
    elif(replay == 'yes' or replay == '1'):
        print('                                  ')
        print('                                  ')
        print('                                  ')
        continue
    else:
        print('잘못 입력하셨으니 종료합니다.')
        break
# ************************************재시작*************************************** #
```
